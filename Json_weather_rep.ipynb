```python
import json
import requests
url="https://api.openweathermap.org/data/2.5/forecast/daily?q=vijayawada,in&cnt=10&mode=json&units=metric&APPID=2d80cf7142a085e6c34f383205d35118"
resp=requests.get(url)
resp
```

```python
d=resp.content
d
```

```python
json.loads(d) ## json.dumps -> converts a python obj to String Obj
# json.loads -> converts a string obj to Python Obj
```

```python
weather_list=d['list']
maxtemp_list=[] #to collect max temps of all days
mintemp_list=[] #to collect min temps of all days
```

```python
for each in weather_list: #I want to find out  min, max temp on that date and find out highest temp and lowest out of all days
    date=each["dt"]
    maxtemp=each['temp']['max']
    mintemp=each['temp']['min']
    maxtemp_list.append(maxtemp)
    mintemp_list.append(mintemp)
    print("Date: ",date ,"maxtemp: ",maxtemp, "mintemp: ",mintemp)
highest_temp1=max(maxtemp_list) #find out highest temp and lowest out of all days
lowest_temp1=min(mintemp_list)
print("Highest temp and lowest out of all days: ",highest_temp1,'C',"and",lowest_temp1,'C')
```
